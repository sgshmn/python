# 시각화

- 많은 데이터를 그래프로 보여주기
- 도구 : matplotlib(그래프), seaborn(그래프), folium(지도시각화) 등
    - matplotlib.pyplot : 기본적인 시각화 방법 제공
    - matplotlib.pylib : 많은 시각화 방법 제공 (numpy와 연계)
    - seaborn : matplotlib을 보조해서 다양한 시각화 방법 제공
- 구글차트 같은 것도 많이 쓴다

- matplolib 사용

```python
import matplotlib.pyplot as plt

# 출력 결과를 브라우저에 바로 출력 (tk 등을 사용하면 새 창을 열어서 출력)
# tk : tkinter 라이브러리 (파이썬의 기본 GUI 라이브러리)
%matplotlib inline
```

- plot() : 선 그래프 그리기
    - x축에 맞는 y축의 값이 만나는 지점을 선으로 연결

```python
y = [10, 34, 20]

plt.plot(y)

# 결과를 출력
plt.show()
```

```python
x = [1, 2, 3]
y = [10, 34, 20]

plt.plot(x, y)
plt.show()
```

- 선의 스타일 변경 : ls (line style)
    - "-" : 선 (얘는 기본값 default)
    - "--" : 뚝뚝 끊어지는 선 (파선, dashed line)
    - "-." : 대시, 점으로 표시되는 선 (일점쇄선, dash-dot line)
    - "." : 점선 (dotted line)

```python
plt.plot(x, y, ls="-.")
plt.show()
```

- 마커 표시 : x축과 y축이 만나는 위치를 표시하는 점 (marker)
    - "." : 점
    - "o" : 원
    - "v" : 삼각형
    - "s" : 사각형
    - "p" : 오각형
    - "D" : 다이아몬드
    

```python
plt.plot(x, y, ls="-", marker="h")
plt.show()
```

- 선의 두께 : lw (line width)

```python
plt.plot(x, y, ls="-", marker="h", lw=10)
plt.show()
```

- 선색깔 : c (color)
    - r : red
    - g : green
    - b : blue
    - k : black
    - m : magenta
    - y : yellow
    - c : cyan
    - p : pink
    - 등등
    - 위에서 안 되는 것도 있다
    - 그냥 약자 안쓰고 원래대로 써도 되니까 당황하지 말자

```python
plt.plot(x, y, ls="-", marker="o", lw=2, c="pink")
plt.show()
```

- 마커의 색깔 : mfc (marker face color)

```python
plt.plot(x, y, ls="-", marker="o", lw=2, c="pink", mfc='b')
plt.show()
```

- 마커 테두리 색깔 : mec (marker edge color)

```python
plt.plot(x, y, ls="-", marker="o", lw=2, c="pink", mfc='y', mec = 'k')
plt.show()
```

- 마커의 크기 : ms (marker size)

```python
plt.plot(x, y, ls="-", marker="o", lw=2, c="pink", mfc='y', mec = 'k', ms=10)
plt.show()
```

- 마커의 테두리 두께 : mew(marker edge width)

```python
plt.plot(x, y, ls="-", marker="o", lw=2, c="pink", mfc='y', mec = 'k', ms=10, mew=3)
plt.show()
```

```python
x = range(7)
y = [1, 4, 5, 8, 9, 5, 3]

plt.plot(x, y)
plt.show()
```

## 실습

- 변경하기

```python
plt.plot(x, y, ls="--", marker="o", lw=5, mfc='r', mec = 'g', ms=15, mew=5)
plt.show()
```

- 그래프 설정 함수

- xlim(), ylim() : 축의 범위를 설정하는 함수

```python
plt.plot(x, y, ls="--", marker="o", lw=5, mfc='r', mec = 'g', ms=15, mew=5)

plt.xlim(-5, 10)
plt.ylim(-2, 13)

plt.show()
```

- xticks(), yticks() : 축 눈금 설정
    - [ ]로 쓰면 축의 값을 출력하지 않는다
    - 이미지 출력할 때 주로 사용한다

```python
plt.plot(x, y, ls="--", marker="o", lw=5, mfc='r', mec = 'g', ms=15, mew=5)

plt.xticks([0, 3, 6])
plt.yticks([1, 5, 9])

plt.show()
```

- grid() : 격차 표시 여부
- 정확하게 값을 읽기 위해

```python
plt.plot(x, y, ls="--", marker="o", lw=5, mfc='r', mec = 'g', ms=15, mew=5)

plt.xlim([-5, 10])
plt.ylim([-2, 13])

plt.grid()

plt.show()
```

- 플롯 여러개 띄우기

```python
x = [1, 2, 3, 4]
y1 = [10, 30, 32, 24]
y2 = [40, 33, 12, 44]
y3 = [34, 20, 12, 16]


# show를 실행하지 않으면 겹쳐 그려진다
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
```

```python
plt.plot(x, y1)
plt.show()
plt.plot(x, y2)
plt.show()
plt.plot(x, y3)
```

- 범례표시하기
- 그래프가 뭐를 그린건지 표시
- label 매개변수로 설정하고 plt.legend()함수로 호출한다

```python
plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')
plt.plot(x, y3, label='y3')

# best 매개변수를 설정하지 않으면 알아서 배치한다
# 배치는 전체를 상하좌우 중앙으로 9개로 나눈 것 중 하나를 선택할 수 있다
plt.legend()

plt.show()
```

- xlabel(), ylabel(), title()
- x축이름, y축이름, 그래프이름

```python
plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')
plt.plot(x, y3, label='y3')

# 한글은 지원하지 않으므로 폰트를 따로 설정해야한다
plt.xlabel("xxx")
plt.ylabel("yyy")
plt.title("graphs")
plt.show()
```

- 그래프의 종류

- bar : 막대그래프

```python
x = [1, 2, 3, 4]
y = [10, 30, 32, 24]

# color : 색깔 설정 (순서대로 설정)
plt.bar(x,y, color=["gold", "b", "#FF0000", "green"])
plt.show()
```

- barh : 누워있는 막대그래프

```python
plt.barh(x,y, color=["gold", "b", "#FF0000", "green"])
plt.show()
```

- 원그래프 (pie chart)
    - 값들의 합이 1이 되는 데이터에 사용

```python
x = [1, 4, 3, 2]

plt.pie(x, labels=["A 1", "B 4", "C 3", "D 2"])
plt.show()
```
