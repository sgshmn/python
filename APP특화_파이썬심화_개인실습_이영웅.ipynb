[ 문제 2 ] 두 개의 정수와 연산자를 입력받은 후, 입력한 연산자에 따라 연산결과가 나오는 프로그램을 함수를 활용하여 작성하시오.

- inputNum() : 두 개의 정수와 연산자를 입력받아 반환하는 함수
- calNum(n1, n2, op) : 두 개의 정수와 연산자를 매개변수로 받아서 결과를 반환하는 함수



<font size=5 color=red><b>[ 문제 2 답안 ]</b></font>

```python
def inputNum():
    n1 = int(input("첫 번째 정수 입력 : "))
    n2 = int(input("두 번째 정수 입력 : "))
    op = input("연산자 입력 : ")
    return n1, n2, op

def calNum(n1, n2, op):
    if op == '+':
        return n1 + n2
    elif op == '-':
        return n1 - n2
    elif op == '*':
        return n1 * n2
    elif op == '/':
        return n1 / n2

print("==== 사칙연산 (+, -, *, /) ====")
n1, n2, op = inputNum()
print(f"연산 결과 : {calNum(n1, n2, op)}")
```

[ 문제 3 ] 두개의 정수를 입력인수로 받아 num1부터 num2까지의 숫자 중 소수를 출력하는 기능을 구현하시오 (소수 : 약수가 1과 자기 자신 밖에 없는 수).
- CalClass1 클래스에 두개의 정수를 입력받아서 num1, num2에 저장하는 inputNum() 함수 구현
- CalClass2 클래스에 두개의 정수를 입력받아서 소수를 판단하고 출력하는 getPrimeNumber() 함수 구현
- CalClass 클래스는 CalClass1와 CalClass2를 상속받음
- CalClass 클래스에서 inputNum()와 getPrimeNumber()를 호출하여 아래와 같이 동작되도록 구현



<font size=5 color=red><b>[ 문제 3 답안 ]</b></font>

```python
class CalClass1():
    
    num1 = 0
    num2 = 0
    def inputNum(s):
        s.num1 = int(input("시작 값 입력 (1보다 큰 값) : "))
        s.num2 = int(input("끝 값 입력 : "))

class CalClass2():
    
    def getPrimeNumber(s, num1, num2):
        print(f"{num1}~{num2}까지의 소수 : ", end="")
        for i in range(num1, num2 + 1):
            mmm = 0
            for j in range(2, i // 2 + 1):
                if i % j == 0:
                    mmm += 1
            if mmm == 0:
                print(i, end=" ")
            
            
class CalClass(CalClass1, CalClass2):
    pass

ccc = CalClass()

ccc.inputNum()
ccc.getPrimeNumber(ccc.num1, ccc.num2)
```
