- apply() : 기존 데이터를 분해한 뒤 함수를 호출하고 결과를 병합하는 함수 --> 재구조화 함수

```python
# 과목의 최대점수와 최소점수의 차이를 구해서 차이 컬럼에 추가
# 최대점수와 최소점수의 차이를 구해서 결과를 반환하는 함수

# (2) apply() 함수에서 받은 값을 이용해서 최대값과 최소값으리 차이를 구하고 반환
def min_max(row):
    return row.max() - row.min()

# (1) apply() 함수는 score의 데이터를 한 줄 씩(axis=1) min_max 함수로 넘긴다
# (3) min_max의 반환 값을 병합
score["차이"] = score.loc[:4,"1반":"4반"].apply(min_max, axis=1)

score
```

```python
# 등급 컬럼을 만들고 차이가 50보다 크면 A, 그렇지 않으면 B로 저장
def ab(e):
    return 'A' if e > 50 else 'B'

score["등급"] = score.loc[:4,"차이"].apply(ab)
score
```

- cut() : 데이터를 여러 구간으로 쪼개서 각 구간에 값을 부여
    - 1차원 데이터만 적용
    - 나눈 구간의 수와 라벨의 수는 같아야한다
    
    - cut(데이터, 구간수, 구간의 라벨)

```python
import numpy as np
import pandas as pd

df = pd.DataFrame({"col":np.random.randn(10)}) # randn 정규분포

df
```

```python
cut_df = pd.cut(df['col'], 4, labels=["A", "B", "C", "D"])

cut_df
```

- cut (데이터, 구간의 범위, 구간의 라벨) 이렇게도 쓴다
    - right = False 이면 구간의 범위에서 왼쪽 수는 포함하고, 오른쪽 수는 포함하지 않는다
    - 만약 right = True로 설정하면 반대로 된다 근데 이게 default이다

```python
age = [1, 13, 33, 74, 23, 27, 40, 3, 57, 61]

# 1번 구간 : 1~19
# 2번 구간 : 20~39
# 3번 구간 : 40~59
# 4번 구간 : 60~74
bins = [1, 20, 40, 60, 75]
labels = ["유아청소년", "청년", "중년", "노년"]

age_cut = pd.cut(age, bins, labels = labels, right=False)

age_cut
```

## 실습

- weight = [30, 90, 67, 45, 94, 66, 45, 56, 75, 89]에서 50미만은 저체중, 50이상 70미만 정상, 70이상 80미만 고체중, 80이상 비만

```python
weight = [30, 90, 67, 45, 94, 66, 45, 56, 75, 89]
flags = [1, 50, 70, 80, 300]
names = ["저체중", "정상", "고체중", "비만"]

weight_cut = pd.cut(weight, flags, labels = names, right=False)

weight_cut
```

```python
weight_cut.value_counts()
```

```python
weight_cut.categories
```

- qcut() : 데이터의 개수가 동일한 범위를 찾아서 구분한다

```python
weight_cut2.value_counts()
```

- pivot_table() : 많은 양의 데이터에서 필요한 자료만 정리해서 표로 작성한다
- groupby() : 원하는 컬럼을 기준으로 데이터를 모은다

```python
df = pd.DataFrame({
    "년도":["2015", "2017", "2015", "2017"],
    "지역":["광주", "광주", "전남", "전남"],
    "사과":[500, 450, 300, 600],
    "배":[780, 650, 530, 550],
    "딸기":[200, 250, 225, 250]
})

df
```

```python
# 년도를 기준으로 모아보자

df_group = df.groupby(["년도"])
df_group
```

```python
df_group.mean()
```

```python
# 지역을 기준으로 모아보자
df_group = df.groupby(["지역"])
df_group
```

```python
df_group.mean()
```

- 년도와 지역을 기준으로 모아보자

```python
df_group = df.groupby(["년도", "지역"])
df_group
```

```python
df_group.mean()
```

```python
# 기본 값으로 평균을 사용한다
# 컬럼을 이름 순으로 재정렬한다

df_pivot = pd.pivot_table(df, index=["년도", "지역"])
df_pivot
```

```python
# 피벗테이블의 연상방법 변경 aggfunc
df_pivot = pd.pivot_table(df, index=["지역"], aggfunc=sum)
df_pivot
```

```python
# 딸기 컬럼만 적용하기
# values : 연산 대상 컬럼
# index : 분류 기준
df_pivot = pd.pivot_table(df, index=["년도"], values="딸기", aggfunc=sum)
df_pivot
```

```python
# 피벗 테이블의 데이터에 접근하는 방법
# 피벗 테이블은 DataFrame이다

df_pivot.loc["2017"]
```
