- apply() :          -->  

```python
#        
#       

# (2) apply()         
def min_max(row):
    return row.max() - row.min()

# (1) apply()  score    (axis=1) min_max  
# (3) min_max   
score[""] = score.loc[:4,"1":"4"].apply(min_max, axis=1)

score
```

```python
#     50  A,   B 
def ab(e):
    return 'A' if e > 50 else 'B'

score[""] = score.loc[:4,""].apply(ab)
score
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame({"col":np.random.randn(10)}) # randn 

df
```

- cut (,  ,  )  
    - right = False      ,    
    -  right = True      default

```python
age = [1, 13, 33, 74, 23, 27, 40, 3, 57, 61]

# 1  : 1~19
# 2  : 20~39
# 3  : 40~59
# 4  : 60~74
bins = [1, 20, 40, 60, 75]
labels = ["", "", "", ""]

age_cut = pd.cut(age, bins, labels = labels, right=False)

age_cut
```

```python
weight = [30, 90, 67, 45, 94, 66, 45, 56, 75, 89]
flags = [1, 50, 70, 80, 300]
names = ["", "", "", ""]

weight_cut = pd.cut(weight, flags, labels = names, right=False)

weight_cut
```

```python
weight_cut.value_counts()
```

```python
weight_cut.categories
```
