[ 문제 ] 다음은 날씨 데이터 조회 프로그램입니다. 조건에 맞게 프로그램을 작성하시오.

- 프로그램을 시작하면 다음과 같은 메뉴가 출력
- [5]종료 메뉴를 선택하기 전까지 반복 
- [5]종료 메뉴를 선택할 시 프로그램 종료
  


- [1]날씨입력 메뉴를 선택할 시 다음과 같이 날짜, 온도, 습도, 기압, 날씨를 입력 받아 weather 딕셔너리에 추가
- 온도, 습도, 기압은 정수형으로 입력받음

   - ["2022-08-01", 30, 65, 980, "Sunny"]
   - ["2022-08-02", 34, 88, 990, "cloudy"]
   - ["2022-08-03", 30, 90, 1000, "rain"]
    


- [2] 날씨 조회 메뉴를 선택하면 전체 날씨 정보가 표시



- [3] 날씨 수정 메뉴를 선택하면 원하는 날짜를 입력하여 온도, 습도, 기압, 날씨를 수정

  - ["2022-08-03", 30, 90, 900, "cloudy"]
  


- [4] 날씨 삭제 메뉴를 선택하면 원하는 날짜를 입력하여 해당 날짜 데이터 삭제



<font size=5 color=red><b>[ 문제 1 답안 ]</b></font>

```python
def menu():
    return int(input("[1] 날씨 입력 [2] 날씨 조회 [3] 날씨 수정 [4] 날씨 삭제 [5] 종료 >> "))

def equals():
    print('='*70)


weather = {}
data = []
def wein():
    date = input("날짜 입력 (0000-00-00) >> ")
    temp = int(input("온도 입력 >> "))
    humid = int(input("습도 입력 >> "))
    press = int(input("기압 입력 >> "))
    sun = input("날씨 입력 >> ")
    data = [temp, humid, press, sun]
    weather[date] = data
    equals()

def wecheck():
    equals()
    print("날짜\t\t온도\t습도\t기압\t날씨")
    equals()
    for key, value in weather.items():
        print(f"{key}\t", end="")
        for v in value:
            print(f"{v}\t", end="")
        print()
    equals()

def were() :
    date = input("날짜 입력 (0000-00-00) >> ")
    if date not in weather.keys():
        print("그 날짜 데이터는 없습니다")
        return None
    temp = int(input("온도 입력 >> "))
    humid = int(input("습도 입력 >> "))
    press = int(input("기압 입력 >> "))
    sun = input("날씨 입력 >> ")
    data = [temp, humid, press, sun]
    weather[date] = data
    equals()
    
    
def werm():
    equals()
    date = input("날짜 입력 (0000-00-00) >> ")
    del weather[date]
    equals()


s = "="
print(s*20 + "날씨 조회 프로그램" + s*20)

while True:
    ch = menu()   
    if ch == 1:
        wein()
    elif ch == 2:
        wecheck()
    elif ch == 3:
        were()
    elif ch == 4:
        werm()
    elif ch == 5:
        break
```

[ 문제 ] 

- (1) 1과목의 weather 값을 데이터프레임으로 변환하세요.



- (2) 다음 데이터를 데이터프레임으로 만들고 기존 데이터에 concat() 함수를 이용해서 추가해보세요.

  {'2022-08-04': {'온도': 34, '습도': 75, '기압': 980, '날씨': '맑음'},<br>
   '2022-08-05': {'온도': 26, '습도': 90, '기압': 1005, '날씨': '비'}}
   
   

<font size=5 color=red><b>[ 문제 (1) 답안 ]</b></font>

```python
import pandas as pd

dfw = pd.DataFrame(weather, index=["온도", "습도", "기압", "날씨"])

dfw = dfw.T
dfw
```

<font size=5 color=red><b>[ 문제 (2) 답안 ]</b></font>

```python
dfw2 = pd.DataFrame({'2022-08-04': {'온도': 34, '습도': 75, '기압': 980, '날씨': '맑음'},
'2022-08-05': {'온도': 26, '습도': 90, '기압': 1005, '날씨': '비'}})

dfw2 = dfw2.T

dfw3 = pd.concat([dfw, dfw2])

dfw3
```
